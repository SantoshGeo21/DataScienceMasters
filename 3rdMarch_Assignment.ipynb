```python
#Q.1 Ans:-
'''
Below are the five plots that we can plot using the Seaborn library along with its uses:-

1.Line Plot
We use the seaborn line plot mainly to visualize the given data in some time-series form, i.e., in a continuous manner with 
respect to time.

2.Dist plot
We use the seaborn dist plots to plot histograms with the given variables and data as a result.

3.Lmplot
It shows a line representing a linear regression model along with data points on the 2D-space and x and y can be set as the 
horizontal and vertical labels respectively.

4.Countplot
A countplot basically counts the categories and returns a count of their occurrences.It is one of the simplest plots provided 
by the seaborn library.

5.Barplot
A barplot is basically used to aggregate the categorical data according to some methods and by default it’s the mean.
```

```python
#Q.2 Ans:-
import seaborn as sns
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x = "timepoint", y = "signal", data = fmri, hue = 'event', style='region')
```
