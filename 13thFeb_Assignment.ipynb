```python
#Q.4 Ans:-
'''
LookupError is a base class for exceptions that occur when a specified key or index is not found in a 
collection. It is a subclass of the Exception class, which is the base class for all built-in exceptions 
in Python.

Index and Key Error are two subclass of LookupError.

"IndexError" error occurs when an attempt is made to access an item in a list at an index which is out 
of bounds.
'''
# IndexError
l = [1,2,3,3]
l[10]

'''
"KeyError" is raised when a mapping key is accessed and isn't found in the mapping.
'''
# KeyError
d = {1: [3,4,5,6] , "key" :"sudh"}
d["key10"]
```

```python
#Q.5 Ans:-
'''
ImportError is a built-in exception in Python that is raised when an import statement fails to import a 
module or when a module has a problem during its initialization. An ImportError can be raised for a 
variety of reasons, such as a missing module, a syntax error in the module, or a circular import 
dependency.

In Python 3.6 and later, a new exception called ModuleNotFoundError was introduced as a subclass of 
ImportError. ModuleNotFoundError is raised when an import statement fails to find a specified module.
```
