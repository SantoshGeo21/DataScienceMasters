```python
#Q.1 Ans:-
'''
MongoDB is a popular NoSQL database system that is designed to store and manage unstructured data. It uses a document-oriented 
data model, which means data is stored in flexible and hierarchical documents instead of rows and tables.

Non-relational Databases:-

Non-relational databases, or NoSQL databases, are a class of databases that don't rely on a fixed schema like traditional 
relational databases. They are designed to handle unstructured or semi-structured data, such as social media data, sensor data,
or log files. NoSQL databases typically scale horizontally and offer higher availability than traditional databases.

MongoDB over SQL databases:-

-MongoDB is highly and easily scalable.
-Maintaining NoSQL Servers is Less Expensive.
-No Schema or Fixed Data model.
-It allows more dynamic and complex data stuctures.
-Highly suitable for real-time analytics, content management systems, and mobile applications.
```

```python
#Q.2 Ans:-
'''
MangoDB is a polpular NoSQL database having below key features.

Document-oriented data model: MongoDB stores data in flexible, JSON-like documents that can have varying structures, making it 
easier to model complex data.

Dynamic schema: MongoDB does not enforce a strict schema, allowing for dynamic changes in data structure and making it easier 
to handle evolving data.

High performance: MongoDB is designed for high-performance reads and writes, with features like sharding and replica sets that 
allow it to scale horizontally and handle large volumes of data.

Indexing and querying: MongoDB offers powerful indexing and querying capabilities, including support for geospatial queries, 
full-text search, and aggregation pipelines.

Automatic scaling: MongoDB can automatically rebalance data across servers to ensure optimal performance and scalability.

High availability: MongoDB offers features like replica sets and automatic failover to ensure high availability and data 
redundancy.

Flexible deployment: MongoDB can be deployed on-premise or in the cloud, and supports multiple deployment models, including 
standalone, replica set, and sharded clusters.

Easy integration: MongoDB has extensive driver support for multiple programming languages, making it easy to integrate with 
popular programming languages and frameworks.
```

```python
#Q.3 Ans:-

import pymongo

# Connect to DB
client = pymongo.MongoClient("mongodb://localhost:27017/TestDB")

# Create DB
db = client["TestDB"]

# Create Collection
collection = db['Products']
```

```python
#Q.4 Ans:-

d = {'companyName': 'pwskills',
     'product': 'Affordable AI',
     'courseOffered': 'Machine Learning with Deployment'}
rec = collection.insert_one(d)

list_of_records = [
    {'companyName': 'iNeuron',
     'product': 'Affordable AI',
     'courseOffered': 'Machine Learning with Deployment'},
    
    {'companyName': 'iNeuron',
     'product': 'Affordable AI',
     'courseOffered': 'Deep Learning for NLP and Computer vision'},
    
    {'companyName': 'iNeuron',
     'product': 'Master Program',
     'courseOffered': 'Data Science Masters Program'}
]
collection.insert_many(list_of_records)

all_record = collection.find()
for idx, record in enumerate(all_record):
     print(f"{idx}: {record}")


print(collection.find_one())
```

```python
#Q.5 Ans:-

# All the record at once present
all_record = collection.find()

for idx, record in enumerate(all_record):
     print(f"{idx}: {record}")

# Useing filter in find()
for i in collection.find({'companyName': 'iNeuron'}):
    print(i)
```
